## 1.2 Looking at columns with "," separator 

```python
# rating of values in the column

def rating_of_column_values(df, column, separator=', '):
    df['column_2'] = df[column].str.split(separator)
    top_tags_list = list(set().union(*df['column_2'].dropna()))

    top_tags_list_2 = []
    for top_tag in top_tags_list:
        top_tags_list_2.append([top_tag, df['column_2'].dropna().map(set([top_tag]).issubset).fillna(False).value_counts()[True]])

    top_tags_list_2 = sorted(top_tags_list_2, key=lambda x: x[1], reverse=True)
    df = df.drop(['column_2'], axis=1)
    return top_tags_list_2

```

```python
# columns with ', ' (amount and percentage)

for col in rest_df.select_dtypes(include='object').columns:
    rest_with_comma = len(rest_df[rest_df[col].str.contains(', ', na=False)])
    if rest_with_comma > 0:
        print("{0:<20s} {1:>10} {2:>10f}".format(col, rest_with_comma, round(rest_with_comma/rest_df.shape[0]*100, 2)))
```

## 1.3 Fixing the country name

```python
def theUK(country):
    if country in ['England', 'Scotland', 'Wales', 'Northern Ireland']:
        return 'United Kingdom'
    else:
        return country

def theNe(country):
    if country == 'Netherlands':
        return 'The Netherlands'
    else:
        return country

subreg = {
    'Italy': 'South',
    'Spain': 'South',
    'France': 'West',
    'United Kingdom': 'North',
#     'England': 'North',
    'Germany': 'West',
    'Greece': 'South',
    'Portugal': 'South',
    'The Netherlands': 'West',
    'Poland': 'East',
    'Belgium': 'West',
    'Austria': 'West',
    'Sweden': 'North',
    'Czech Republic': 'East',
#     'Scotland': 'North',
    'Ireland': 'North',
    'Denmark': 'North',
#     'Wales': 'North',
    'Croatia': 'South',
    'Romania': 'East',
    'Hungary': 'East',
    'Finland': 'North',
    'Bulgaria': 'East',
    'Slovakia': 'East'
#     'Northern Ireland': 'North'
}
```

```python
def to_get_dummies(df, columns):
    dummies_df = []

    for col in columns:
        supported_df = df[col].str.get_dummies(', ')
        dummies_df.append(supported_df)

    concat_dummies_df = pd.concat(dummies_df, axis=1).fillna(0)
    return concat_dummies_df


ger_dummies_discrete_df = to_get_dummies(ger2_df, discrete_columns)
ger_dummies_discrete_df
# ger_dummies_discrete_df[ger_dummies_discrete_df.notna()]
```

## 2.3 Applying the Jaccard metric

```python
'''
the content-based recommendation
'''
```

```python
# Import functions from scipy
from scipy.spatial.distance import pdist, squareform

def jaccard_top(df, target_index):
    # Calculate all pairwise distances
    jaccard_distances = pdist(df.values, metric='jaccard')

    # Convert the distances to a square matrix
    jaccard_similarity_array = 1 - squareform(jaccard_distances)

    # Wrap the array in a pandas DataFrame
    jaccard_similarity_df = pd.DataFrame(jaccard_similarity_array, index=df.index, columns=df.index)

    # Find the values for the restaurant with target_index
    jaccard_similarity_series = jaccard_similarity_df.loc[target_index]
        
    # Sort these values from highest to lowest
    ordered_similarities = jaccard_similarity_series.sort_values(ascending=False)

    return ordered_similarities
```

```python
def top_of_similarity(similar_df, threshold):
    similar_df_head = similar_df[similar_df.values >= threshold]
    similar_df_head2 = pd.DataFrame(similar_df_head)
    similar_df_head2.columns = ['similarity_score']
    
    return similar_df_head2
```

```python
similar_df = jaccard_top(ger_dummies_discrete_df, 'g187309-d10000500')
similar_df_head = top_of_similarity(similar_df, 1)
similar_df_head.join(ger_df.set_index('restaurant_link')).reset_index(drop=True)
```
