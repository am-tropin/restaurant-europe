## 1.2 Looking at columns with "," separator 

```python
# rating of values in the column

def rating_of_column_values(df, column, separator=', '):
    df['column_2'] = df[column].str.split(separator)
    top_tags_list = list(set().union(*df['column_2'].dropna()))

    top_tags_list_2 = []
    for top_tag in top_tags_list:
        top_tags_list_2.append([top_tag, df['column_2'].dropna().map(set([top_tag]).issubset).fillna(False).value_counts()[True]])

    top_tags_list_2 = sorted(top_tags_list_2, key=lambda x: x[1], reverse=True)
    df = df.drop(['column_2'], axis=1)
    return top_tags_list_2

```

```python
# columns with ', ' (amount and percentage)

for col in rest_df.select_dtypes(include='object').columns:
    rest_with_comma = len(rest_df[rest_df[col].str.contains(', ', na=False)])
    if rest_with_comma > 0:
        print("{0:<20s} {1:>10} {2:>10f}".format(col, rest_with_comma, round(rest_with_comma/rest_df.shape[0]*100, 2)))
```

## 1.3 Fixing the country name

```python
def theUK(country):
    if country in ['England', 'Scotland', 'Wales', 'Northern Ireland']:
        return 'United Kingdom'
    else:
        return country

def theNe(country):
    if country == 'Netherlands':
        return 'The Netherlands'
    else:
        return country

subreg = {
    'Italy': 'South',
    'Spain': 'South',
    'France': 'West',
    'United Kingdom': 'North',
#     'England': 'North',
    'Germany': 'West',
    'Greece': 'South',
    'Portugal': 'South',
    'The Netherlands': 'West',
    'Poland': 'East',
    'Belgium': 'West',
    'Austria': 'West',
    'Sweden': 'North',
    'Czech Republic': 'East',
#     'Scotland': 'North',
    'Ireland': 'North',
    'Denmark': 'North',
#     'Wales': 'North',
    'Croatia': 'South',
    'Romania': 'East',
    'Hungary': 'East',
    'Finland': 'North',
    'Bulgaria': 'East',
    'Slovakia': 'East'
#     'Northern Ireland': 'North'
}
```

```python
def compare_by_feature(df, features_col, feature):

    df2 = df[
        ['restaurant_link','total_reviews_count','avg_rating',features_col]
    ][df['total_reviews_count'] >= 20]
    df2['feature_flg'] = df2[features_col].apply(lambda x: 1 if feature in str(x) else 0)

    df2['summary_score'] = df2['total_reviews_count'] * df2['avg_rating']

    df2_agg = df2.groupby('feature_flg').agg({'restaurant_link': 'count', 'summary_score': 'sum', 'total_reviews_count': 'sum'})
    df2_agg['total_avg_rating'] = df2_agg['summary_score'] / df2_agg['total_reviews_count']

    return df2_agg[['restaurant_link', 'total_avg_rating']]

```

```python
bavaria_df2_agg = compare_by_feature(bavaria_df, 'features', 'Reservations')
bavaria_df2_agg
```

```python
bavaria_df2_agg = compare_by_feature(bavaria_df, 'top_tags', 'Cheap Eats')
bavaria_df2_agg
```

```python
bavaria_df2_agg = compare_by_feature(bavaria_df, 'meals', 'Lunch')
bavaria_df2_agg
```

```python
feature_columns = ['meals', 'cuisines', 'special_diets', 'features']

for col in feature_columns:
    list_of_features = [f[0] for f in rating_of_column_values(bavaria_df, col, separator=', ') if f[1] >= 20]
    for feature in list_of_features:
        print()
        print(col, '-', feature)
        print(compare_by_feature(bavaria_df, col, feature))
```
